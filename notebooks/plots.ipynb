# Plots

```python
import pandas as pd 
import numpy as np 
import re
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
import seaborn as sns
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})
```

```python
mob_male = pd.read_csv('../data/sim_mob_male_m1.csv')
mob_female = pd.read_csv('../data/sim_mob_female_m1.csv')
```

```python
mob_male.loc[mob_male['q']==3,:].head()
```

```python
mob_female.loc[mob_female['q']==1,:].head()
```

```python
mob_male.info()
```

```python
sns.kdeplot(mob_male.loc[mob_male.q==1, 'diff'], label='Q1', shade=True, alpha=0.4)
sns.kdeplot(mob_male.loc[mob_male.q==2, 'diff'], label='Q2', shade=True, alpha=0.2)
sns.kdeplot(mob_male.loc[mob_male.q==3, 'diff'], label='Q3', shade=True, alpha=0.2)
sns.kdeplot(mob_male.loc[mob_male.q==4, 'diff'], label='Q4', shade=True, alpha=0.2)
# sns.kdeplot(data.split_frac[data.gender=='W'], label='women', shade=True)
plt.xlabel('Life expectancy first difference \n One standard deviation increase in income mobility');
plt.ylabel('Density');
plt.ylim([0, 12])
plt.legend(ncol=4, loc="upper center");
plt.axvline(x=0.0, color='grey', linewidth=1, linestyle='--')
plt.savefig('plots/first_difference_male_mob_m1.pdf')
```

```python
sns.kdeplot(mob_female.loc[mob_female.q==1, 'diff'], label='Q1', shade=True, alpha=0.2)
sns.kdeplot(mob_female.loc[mob_female.q==2, 'diff'], label='Q2', shade=True, alpha=0.2)
sns.kdeplot(mob_female.loc[mob_female.q==3, 'diff'], label='Q3', shade=True, alpha=0.2)
sns.kdeplot(mob_female.loc[mob_female.q==4, 'diff'], label='Q4', shade=True, alpha=0.2)
plt.xlabel('Life expectancy first difference \n One standard deviation increase in income mobility');
plt.ylabel('Density');
plt.ylim([0, 12])
plt.legend(ncol=4, loc="upper center");
plt.axvline(x=0.0, color='grey', linewidth=1, linestyle='--')
plt.savefig('plots/first_difference_female_mob_m1.pdf')
```
